```python
def factorial(num):
    if num ==1:
        return 1
    return num*factorial(num-1)
```

```python
a = factorial(5)
print(a)
```

```python
# Recursion
# A function calling itself is called as a recursive function
# type your code here
# This goes into an infinite loop

# def countdown(n):
#     print(n)
#     countdown(n-1)

# countdown(5)
```

```python
# Recursion
# A function calling itself



def countdown(n):
    if n<=0:
        print("Blast Off!!!!!")
    else:
        print(n)
        countdown(n-1)

countdown(5)
```
